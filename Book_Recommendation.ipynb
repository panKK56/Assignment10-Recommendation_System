## Calculating cosine similarity between users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim=1-pairwise_distances(new_books.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
# set index and column names tu user ids
user_sim_df.index=books.userId.unique()
user_sim_df.columns=books.userId.unique()
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df
```

# Most Similar Users

```python
print(user_sim_df.idxmax(axis=1)[1348])
print(user_sim_df.max(axis=1).sort_values(ascending=False).head())
```
